```python
import pandas as pd
```

```python

# ======================
# FUNÇÕES DO PIPELINE
# ======================

def pipeline(df):

    def criar_coluna_dobro(df):
        df["idade_dobro"] = df["idade"] * 2
        return df

    def remover_nulos(df):
        return df.dropna()

    df = remover_nulos(df)
    df = criar_coluna_dobro(df)

    return df
```

```python
# ======================
# TESTE NO NOTEBOOK
# ======================

df = pd.DataFrame({
    "nome": ["Lucas", "Maria", None],
    "idade": [23, 30, 40]
})

resultado = pipeline(df)

# Testes
assert len(resultado) == 2
assert "idade_dobro" in resultado.columns
assert resultado.iloc[0]["idade_dobro"] == 46
assert resultado.iloc[1]["idade_dobro"] == 60

print("✅ Todos os testes passaram!")
```
